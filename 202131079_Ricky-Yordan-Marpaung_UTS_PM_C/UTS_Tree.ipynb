## PREPROCESSING

```python
# Menggunakan LabelEncoder untuk mengubah data kategorikal menjadi numerik
from sklearn.preprocessing import LabelEncoder
```

```python
for col in ['map', 'round_winner']:
        label_encoder = LabelEncoder()
        data[col] = label_encoder.fit_transform(data[col])
```

```python
data
```

```python
# Memisahkan fitur (x) dan target (y)
x = data.drop('round_winner', axis = 1)
y = data['round_winner']
```

```python
x
```

```python
y
```

```python
# Membagi data menjadi data pelatihan dan data uji
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Modelling

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier(max_depth=3)
modelDT.fit(x_train, y_train)
```

```python
# Melakukan prediksi pada data uji
hasil_prediksi_dt = modelDT.predict(x_test)
```

```python
# Menampilkan akurasi model
from sklearn.metrics import accuracy_score
print('Akurasi Model ', accuracy_score(y_test, hasil_prediksi_dt) * 100)
```

```python
# Menampilkan visualisasi pohon keputusan
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(40, 20))
plot_tree(modelDT, fontsize=30)
plt.show()
```
